#### Remove Null Values from the data
**Null values affects the performance and accuracy** of the machine learning model and therefore it is important to identify them and deal with them accordingly.

```python
data.isnull().sum() 
```

#### Replace null values with mean values for respective columns 
We replaced columns with respective values so that we can still use those rows in our data

```python
columns_with_na = ['education','cigsPerDay','BPMeds','totChol','BMI','heartRate','glucose']

for i in columns_with_na: 
    data[i].fillna(data[i].mean(),inplace=True)
    
data.isnull().sum() 
```

#### We use the information above to remove highly correlated features

```python
highly_correlated_features = ['currentSmoker','diaBP','prevalentHyp','diabetes']
data.drop(highly_correlated_features,axis=1,inplace=True)
```

#### Training on the dataset

```python
X = data.drop('TenYearCHD',axis=1)
y = data['TenYearCHD']
```

#### Using SMOTE to balance the data


```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_ros, y_ros = smote.fit_resample(X, y)
ros_chd_plot=y_ros.value_counts().plot(kind='bar')
plt.show()
```

#### Train vs Test Splits

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_ros,y_ros,test_size=0.2,random_state=42)
X_train.head()
```

#### Scale the data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=pd.DataFrame(sc.fit_transform(X_train))
X_test=pd.DataFrame(sc.transform(X_test))
X_train.head()
```

#### Reset colmns (Scaling removed column indexes)

```python
X_train.columns= X.columns   
X_test.columns= X.columns

y_train.index= X_train.index
y_test.index= X_test.index

X_train.head()
```

#### Create a Logistic regression model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
 pred = model.predict_proba(X_test)
```
